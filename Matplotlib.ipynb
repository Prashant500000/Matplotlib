# Bar plot

```python
import matplotlib.pyplot as plt
```

```python
x = ["python", "c", "c++", "java"]
y = [85, 60, 70, 80]
plt.bar(x,y)
plt.show()
```

```python
x = ["python", "c", "c++", "java"]
y = [85, 60, 70, 80]
plt.xlabel("language")
plt.ylabel("no.")
plt.title("wscube")
plt.bar(x,y)
plt.show()
```

```python
x = ["python", "c", "c++", "java"]
y = [85, 60, 70, 80]
plt.xlabel("language", fontsize = 10)
plt.ylabel("no.", fontsize = 10 )
plt.title("wscube",fontsize = 10 )
plt.bar(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
x = ["python", "c", "c++", "java"]
y = [85, 60, 70, 80]
plt.xlabel("language", fontsize = 25)
plt.ylabel("no.", fontsize = 25 )
plt.title("wscube",fontsize = 25 )
plt.bar(x,y)
plt.show()
```

```python
x = ["python", "c", "c++", "java"]
y = [85, 60, 70, 80]
plt.xlabel("language", fontsize = 10)
plt.ylabel("no.", fontsize = 10 )
plt.title("wscube",fontsize = 10 )
c = ["r", "y", "g", "b"]
plt.bar(x,y, width = 0.4, color = c)
plt.show()
```

```python
x = ["python", "c", "c++", "java"]
y = [85, 60, 70, 80]
plt.xlabel("language", fontsize = 10)
plt.ylabel("no.", fontsize = 10 )
plt.title("wscube",fontsize = 10 )
c = ["r", "y", "g", "b"]
plt.bar(x,y, width = 0.4, color = c , edgecolor = "r", linewidth = 5, linestyle = ":", alpha=0.4)
plt.show()
```

```python
x = ["python", "c", "c++", "java"]
y = [85, 60, 70, 80]
plt.xlabel("language", fontsize = 10)
plt.ylabel("no.", fontsize = 10 )
plt.title("wscube",fontsize = 10 )
c = ["r", "y", "g", "b"]
plt.bar(x,y, width = 0.4, color = c , edgecolor = "r", label = "Prashant")
plt.legend()
plt.show()
```

# Scatter plot

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6,7]
y = [2,4,6,3,5,7,8]
plt.scatter(x,y)
plt.show()
```

```python
x = [1,2,3,4,5,6,7]
y = [2,4,6,3,5,7,8]
plt.scatter(x,y)
plt.title("scatter plot")
plt.show()
```

```python
x = [1,2,3,4,5,6,7]
y = [2,4,6,3,5,7,8]
plt.scatter(x,y)
plt.title("scatter plot", fontsize = 15)
plt.show()
```

```python
x = [1,2,3,4,5,6,7]
y = [2,4,6,3,5,7,8]
plt.scatter(x,y)
plt.title("scatter plot", fontsize = 15)
plt.xlabel("Day")
plt.ylabel("No.")
plt.show()
```
